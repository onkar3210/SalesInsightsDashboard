```python
file_path = r'C:\Users\Onkar\Desktop\Data Analysis 2024\Power Bi\New folder\vikas-0731-global-super-store\global_superstore_2016.xlsx'
orders_df = pd.read_excel(file_path, sheet_name='Orders')
returns_df = pd.read_excel(file_path, sheet_name='Returns')
```

```python
location_df=orders_df[['City','State','Country','Region','Market']]
```

```python
subset_columns = ["City", "State", "Country", "Region", "Market"]
location_df=location_df.drop_duplicates(subset=subset_columns)
```

```python
location_df.head()
```

```python
location_df.reset_index(inplace=True)
```

```python
# Generate unique IDs for each location in the location_df DataFrame
location_df['LocationID'] = range(1, len(location_df) + 1)
```

```python
location_df.head()
```

```python
# Add a foreign key column in the Orders_df to reference the 'LocationID' from the location table
orders_df['LocationID'] = None  # Initialize the column with None
```

```python
# Iterate over each row in Orders_df and match location information to get the corresponding LocationID
for index, order_row in orders_df.iterrows():
    # Match location information to find the corresponding LocationID in location_df
    matching_location = location_df[
        (location_df['City'] == order_row['City']) &
        (location_df['State'] == order_row['State']) &
        (location_df['Country'] == order_row['Country']) &
        (location_df['Region'] == order_row['Region']) &
        (location_df['Market'] == order_row['Market'])
    ]
    # If a match is found, update the LocationID in Orders_df
    if not matching_location.empty:
        orders_df.at[index, 'LocationID'] = matching_location['LocationID'].iloc[0]
```

```python
orders_df.head()
```

```python
orders_df.drop(columns=["City", "State", "Country", "Region", "Market"], inplace=True)
```

```python
# Generate unique IDs for product subcategories and product categories
orders_df['SubcategoryID'] = orders_df['Sub-Category'].astype('category').cat.codes + 1
orders_df['CategoryID'] = orders_df['Category'].astype('category').cat.codes + 1

# Create product subcategory table
product_subcategory_df = orders_df[['Sub-Category', 'SubcategoryID', 'CategoryID']].drop_duplicates().reset_index(drop=True)
product_subcategory_df.columns = ['ProductSubcategoryName', 'ProductSubcategoryKey', 'ProductCategoryKey']

# Create product category table
product_category_df = orders_df[['Category', 'CategoryID']].drop_duplicates().reset_index(drop=True)
product_category_df.columns = ['ProductCategoryName', 'ProductCategoryKey']

# Create product table
product_df = orders_df[['Product Name', 'Product ID', 'SubcategoryID']].drop_duplicates().reset_index(drop=True)
product_df.columns = ['ProductName', 'ProductKey', 'ProductSubcategoryKey']
```

```python
customers_df = orders_df[['Customer ID', 'Customer Name', 'Segment']].drop_duplicates().reset_index(drop=True)
```

```python
# Get the unique Order_id values from the orders_df DataFrame
unique_order_ids = orders_df['Order ID'].unique()

# Keep only the rows in returns_df where Order_id is present in unique_order_ids
filtered_returns_df = returns_df[returns_df['Order ID'].isin(unique_order_ids)]
```
